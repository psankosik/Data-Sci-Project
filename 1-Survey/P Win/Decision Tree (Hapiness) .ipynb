# Decision Tree Approach

Use Decision Tree approach to analyze employee hapiness. I use happy or unhappy as label in this attempt. Happy person will be the upper 66.67th percentile, while unhappy person will be bottom 33.33th percentile.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
rawdata = pd.read_excel("dEX Team Survey.xlsx") 
```

```python
rawdata.head()
```

### Data Wrangling

```python
data = rawdata.drop(["ID","Start time", "Completion time", "Email","Name",'Are you fully empowered and accountable for the work that you do?','Hypothetically, if you were to quit tomorrow, what would your reason be? (Please select one)','Total Score'], axis=1)
```

```python
data.head()
```

```python
data['How happy are you at work?'].describe()
```

```python
#Finding percentile
p = np.percentile(data['How happy are you at work?'], 100/3) 
print(p)
```

```python
#Create label
data['Happy or Not'] = data['How happy are you at work?']>6.01
```

```python
data.head()
```

### Machine Learning

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
from IPython.display import Image  
from sklearn.externals.six import StringIO  
from sklearn.tree import export_graphviz
import pydot 
```

```python
features = list(data.columns[1:-1])
print(len(features))
classname = ['UnHappy','Happy']
#features
```

```python
dot_data = StringIO()  
XX = data.drop(['How happy are you at work?','Happy or Not'],axis=1)
yy = data['Happy or Not']


dtree1 = DecisionTreeClassifier()
dtree1.fit(XX,yy)

export_graphviz(dtree1,class_names=classname, out_file=dot_data,feature_names=features,filled=True,rounded=True)

graph = pydot.graph_from_dot_data(dot_data.getvalue())  
Image(graph[0].create_png())  
```
